```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import expon
%matplotlib inline
%config InlineBackend.figure_format='retina'

xmin,xmax,N=1,10,1000
x = np.random.uniform(xmin,xmax,N)
px=(1/(xmax-xmin))
y=np.exp(x) #This is the issue, working out a way to calculate the exponential distribution with numpy
#py=px*y
plt.hist(x, bins=20, density=True, edgecolor='black')
#plt.hist(py, density=True,color='red')
plt.title("Histogram of 1000 random samples")
plt.xlabel("x")
plt.ylabel("p(x)")

xmean=np.mean(x)
xmedian=np.median(x)
ymean=np.mean(y)
ymedian=np.median(y)
print(xmean,xmedian)
print(ymean,ymedian)
```
